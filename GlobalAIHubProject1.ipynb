```python
def calculateLetterGrade(score):
  if score>=80:
    return "AA"
    print("Harf Notu: AA")
  elif score>=70:
    return "BA"
    print("Harf Notu: BA")
  elif score>=60:
    return "BB"
    print("Harf Notu: BB")
  elif score>=50:
    return "BC"
    print("Harf Notu: BC")
  elif score>=40:
    return "CC"
    print("Harf Notu: CC")
  elif score>=30:
    return "CD"
    print("Harf Notu: CD")
  elif score>=20:
    return "DD"
    print("Harf Notu: DD")
  elif score<20:
    return "FF"
    print("Harf Notu: FF")

def calculate_pass_fail(letter_grade):
    if letter_grade == "FF": 
      return "FAIL"
    return "PASS"



class student():
  def __init__(self, first_name, last_name, school_number, exam_score, letter_grade, pass_fail):
      self.first_name = first_name
      self.last_name = last_name
      self.school_number = school_number
      self.exam_score = exam_score
      self.letter_grade = letter_grade
      self.pass_fail = pass_fail
      

class GradeBook:
    def __init__(self):
        self.students = []
        
    def studentInsert(self):
        name = input("Please enter student name: ")
        surname = input("Please enter student surname: ") 
        stuNo = int(input("Please enter student number: "))
        grade = int(input("Please enter lesson grade: "))
        letter_grade = calculateLetterGrade(grade)
        pass_fail = calculate_pass_fail(letter_grade)
        self.students.append(student(name,surname,stuNo, grade, letter_grade, pass_fail))
     

        
class App :
    def __init__(self):
        self.grade_book = GradeBook()
    
    def run(self):
        print("Please select the defined process...")
        print("1-)Student Information\n","2-)Dataframe to excel\n","3-)Exit")
        dongu = True
        while dongu == True:
            op = int(input("Select and enter the action you want to do from the menu: "))
            if op == 1:
                print("You have selected Student Information Process..")
                self.grade_book.studentInsert()
            elif op == 2:
                print("You have selected the Dataframe...")
                data = self.grade_book.students
                df = pd.DataFrame([s.__dict__ for s in data])
                writer = pd.ExcelWriter('output.xlsx')
                df.to_excel(writer)
                writer.save()
            elif op == 3:
                print("You have selected the Output Process...")
                print ("Thankyou for using the bank!")
                dongu = False
                exit()

            else:
                print("You made an incorrect transaction selection. Please select a valid transaction!!!") 
def main():
    app = App()
    app.run()

main()
```
